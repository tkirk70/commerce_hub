## Variables

### Costs

```python
# CartRover Testing
crt = 595
# CartRover Monthly
crm = 49
# CommerceHub upgrade (max) if not VAN
chmax = 2250
# Montly charge if not VAN
chmon = 20
# CommerceHub upgrade (min) with VAN KC charges not added.
chmin = 1750
```

### Savings

```python
# Orders from May 21 to May 22
yearly_orders = 9850
# Projecting 50% increase in orders
yearly_orders_projected = 15000
# Total Macy's Sales May 21 to May 22
yearly_sales = 96362.85
# Angela's overtime rate per minute
otr = 17/60
# monthly time spent assuming 75 seconds per order
mor = (yearly_orders/12)*1.25
# monthly time spent assuming 75 seconds per order
mor_proj = (yearly_orders_projected/12)*1.25
# Angela's cost times orders (per month)
ac = otr*mor
# Angela's cost times projected increase in orders (per month)
ac_proj = otr*mor_proj
```

### Assuming 75 seconds labor per order.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,18,250)
y = crm*x+chmax+crt
plt.plot(x, y, '-r', label='Integration')
plt.title('Graph of Maximum Integration Costs by month')
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.legend(loc='upper left')

a = np.linspace(0,18,250)
b = ac*a
plt.plot(a, b, '-b', label='Savings')
# plt.title('Graph of Integration Costs by month')
plt.xlabel('Month')
plt.ylabel('Cost $')
plt.legend(loc='upper left')

# pyplot.figtext(x, y, string)
plt.figtext(0.29, 0.2, f"Macy's orders {yearly_orders} and total sales: ${yearly_sales:,}", fontsize = 11)

plt.grid()
plt.show()
```
